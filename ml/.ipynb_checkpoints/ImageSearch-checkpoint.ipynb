```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import numpy as np
import pandas as pd
import os
from PIL import Image
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
IMG_SIZE = 128
dataset_dir = '/content/drive/MyDrive/MIP_Image/Caltech101dataset/101_ObjectCategories/'
truth_category = 'barrel' 
search_image_name = os.path.join(dataset_dir,truth_category,'image_0001.jpg')
categories_to_search = ['Motorbikes','accordion','airplanes','anchor']
categories_to_search.append(truth_category)
images_to_search = []
for category in categories_to_search:
  image_list = os.listdir(os.path.join(dataset_dir,category))[:10]
  for image in image_list:
    images_to_search.append(os.path.join(dataset_dir,
                                       category,
                                       image))
```

```python
class FeatureExtractor(nn.Module):
  def __init__(self):
    super(FeatureExtractor,self).__init__()
    self.feature_extractor = torchvision.models.vgg19(pretrained=True, progress=False) 
    for param in self.feature_extractor.parameters():
            param.requires_grad = False
        
  def forward(self,input_image):
    features = self.feature_extractor(input_image)
    return features
```

```python
normalization_mean = [0.485, 0.456, 0.406]
normalization_std = [0.229, 0.224, 0.225]
image_transforms = torchvision.transforms.Compose([torchvision.transforms.Resize((IMG_SIZE,IMG_SIZE)),torchvision.transforms.ToTensor(),
                                             torchvision.transforms.Normalize(mean=normalization_mean,
                                                                              std=normalization_std)])
def load_image(image_path,image_transformer):
  image = Image.open(image_path)
  transformed_image = image_transformer(image).unsqueeze(0)
  return transformed_image
```

```python
def calculate_euclidean_distance(feature_vector_1,feature_vector_2):
  return torch.sum(torch.square(torch.subtract(feature_vector_1,feature_vector_2)))**(0.5)
```

```python
def calculate_k_similar_images(input_image,search_list,k,model):
  score_list = []
  input_image_tensor = load_image(input_image,image_transforms).to(device)
  input_image_feature = model(input_image_tensor)
  for image in search_list:
    search_image_tensor = load_image(image,image_transforms).to(device)
    search_image_feature = model(search_image_tensor)
    euclidean_distance = calculate_euclidean_distance(input_image_feature,
                                                     search_image_feature)
    score_list.append({'image_id':image,'score':euclidean_distance.item()})
  
  k_images = sorted(score_list,reverse=False, key=lambda x: x['score'])[:k]
  return k_images 
```

```python
model = FeatureExtractor()
model = model.to(device)
```

```python
similar_images = calculate_k_similar_images(search_image_name,images_to_search,6,model)
```

```python
def plot_images(image_list,rows=5):
  plt.figure(figsize=(15,8))
  for num, x in enumerate(image_list):
    img = Image.open(x['image_id'])
    plt.subplot(rows,6,num+1)
    plt.title(x['image_id'].split(dataset_dir)[1])
    plt.axis('off')
    plt.imshow(img)
```

```python
plot_images(similar_images)
```
